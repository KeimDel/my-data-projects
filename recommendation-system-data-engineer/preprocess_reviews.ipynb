```python
df['ID Numerik'] = df['ID Tempat'].str.extract('(\d+)').astype(int)
```

```python
# Hitung jumlah ID Tempat unik untuk setiap Nama Wisata
duplikat_nama_wisata = df.groupby('Nama Wisata')['ID Tempat'].nunique()

# Ambil Nama Wisata yang punya lebih dari 1 ID Tempat
nama_wisata_duplikat = duplikat_nama_wisata[duplikat_nama_wisata > 1].index

# Ambil baris-baris di df yang termasuk Nama Wisata duplikat tadi
duplicates = df[df['Nama Wisata'].isin(nama_wisata_duplikat)]

# Ambil kombinasi Nama Wisata dan ID Tempat yang unik dari hasil duplikat
kombinasi_duplikat = duplicates[['Nama Wisata', 'ID Tempat']].drop_duplicates()

# Urutkan biar rapi (opsional)
kombinasi_duplikat = kombinasi_duplikat.sort_values('Nama Wisata')

print(kombinasi_duplikat)
```

```python
# Mapping ID Numerik yang perlu digabung: {ID asal: ID target}
id_map = {
    88: 61,
    89: 57,
    91: 56,
    73: 58,
    48: 26,
    108: 26,
    79: 59,
    84: 80
}

# Loop untuk menggabungkan review ke ID target
for old_id, new_id in id_map.items():
    # Ambil data dari ID asal dan ID target
    df_old = df[df['ID Numerik'] == old_id]
    df_new = df[df['ID Numerik'] == new_id]

    # Filter review dari old yang belum ada di new berdasarkan kolom 'nama'
    df_to_add = df_old[~df_old['nama'].isin(df_new['nama'])].copy()

    # Update ID dan nama tempat agar sesuai target
    df_to_add['ID Numerik'] = new_id
    df_to_add['ID Tempat'] = df_new['ID Tempat'].iloc[0] if not df_new.empty else df_to_add['ID Tempat']
    df_to_add['Nama Wisata'] = df_new['Nama Wisata'].iloc[0] if not df_new.empty else df_to_add['Nama Wisata']

    # Gabungkan hasilnya ke df
    df = pd.concat([df, df_to_add], ignore_index=True)

# Hapus semua data dari ID asal
df = df[~df['ID Numerik'].isin(id_map.keys())]
```

```python
df = df.sort_values(by="ID Numerik", ascending=True).reset_index(drop=True)
```

```python
# Buat salinan agar data asli tidak terganggu

# 1. Buat DataFrame unik berdasarkan kombinasi Nama Wisata dan ID Tempat
unique_places = df[['Nama Wisata', 'ID Tempat']].drop_duplicates().reset_index(drop=True)

# 2. Tambahkan penomoran baru dari 1 sampai jumlah tempat unik
unique_places['ID Numerik Baru'] = range(1, len(unique_places) + 1)

# 3. Buat ID Tempat baru dengan format T001, T002, dst.
unique_places['ID Tempat Baru'] = ['T' + str(i).zfill(3) for i in unique_places['ID Numerik Baru']]

# 4. Gabungkan mapping ID baru ke df utama
df = df.merge(unique_places, on=['Nama Wisata', 'ID Tempat'], how='left')

# 5. Gantikan kolom lama dengan yang baru
df['ID Numerik'] = df['ID Numerik Baru']
df['ID Tempat'] = df['ID Tempat Baru']

# 6. Hapus kolom sementara
df = df.drop(columns=['ID Numerik Baru', 'ID Tempat Baru'])

# 7. Urutkan kembali berdasarkan ID Numerik
df = df.sort_values(by='ID Numerik').reset_index(drop=True)
```

```python
# 1. Salin dataframe agar tidak merusak data asli
df_privacy = df.copy()

# 2. Buat mapping dari nama unik ke ID User format U000001
unique_names = df_privacy['nama'].drop_duplicates().reset_index(drop=True)
user_id_map = {name: f'U{str(i+1).zfill(6)}' for i, name in enumerate(unique_names)}

# 3. Ganti kolom 'nama' dengan ID User
df_privacy['ID User'] = df_privacy['nama'].map(user_id_map)

# 4. Hapus kolom nama asli
df_privacy = df_privacy.drop(columns=['nama'])

# 5. Tempatkan ID User di kolom pertama (opsional)
cols = ['ID User'] + [col for col in df_privacy.columns if col != 'ID User']
df_privacy = df_privacy[cols]
```
